**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])#Create axis of the larger graph
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])#Create axes of the smaller graph
ax1.plot(x, z, 'r')
ax2.set_xlim([20, 22])#Set limit of the smaller graph x axis
ax2.set_ylim([30, 50]) #set limit of the smaller graph y axis
ax2.plot(x,  y, 'r')
fig
```
